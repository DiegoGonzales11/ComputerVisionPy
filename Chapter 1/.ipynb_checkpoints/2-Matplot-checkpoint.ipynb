## Plotting Images, points and lines

```python
from PIL import Image
from pylab import *

# read image to array
im = array(Image.open('.\Images\cat.jpg'))

# plot the image
imshow(im)

# some points
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]

# plot the points with red star-markers
plot(x,y, 'r*')

```
