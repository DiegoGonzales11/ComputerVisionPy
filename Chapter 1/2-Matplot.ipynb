# Matplotlib

## Plotting Images, points and lines

```python
from PIL import Image
from pylab import *

# read image to array
im = array(Image.open('.\Images\cat.jpg'))

# plot the image
imshow(im)

# some points
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]

# plot the points with red star-markers
plot(x,y, 'r*')

# line plot connecting the first two points
plot(x[:2], y[:2])

# add title and show the plot
title('Plotting: "cat.jpg"')

# axis('off') #optional
show()
```

## Image contours and histograms
For contours visualization you need grayscale images, because you need a single value for every coordinate [x,y]

```python
from PIL import Image
from pylab import *

# read image to array
im = array(Image.open('./Images/cat.jpg').convert('L'))

# create a new figure
figure()
#don't use colors
gray()
# show contours with origin upper left corner
contour(im, origin='image')
axis('equal')
axis('off')

# histogram
figure()
# have to use the flatten(), because hist need one dimentional array
hist(im.flatten(), 128) 
show()
```
